```python
plt.figure(figsize=(20,30))
plt.plot(df_Canada['date'], df_Canada['total_cases'])
plt.title('Evolution of Confirmed Covid-19 cases', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('cases', fontsize=16)
plt.show()
    
```

```python
# related the variable with the scatterplot 
sns.relplot(x="total_cases",y="total_deaths", data = df_Canada, marker='o')
```

```python
sns.displot(df_Canada['total_cases'], color='c',height=5,aspect=1.2 )
plt.show()
sns.displot(df_Canada['new_cases'], color='c',kde =True,height=5,aspect=1.2)
plt.show()
```
